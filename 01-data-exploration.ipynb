```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


pd.set_option('display.max_columns', None)
file_path = 'data/netflix_titles.csv' 
df = pd.read_csv(file_path)
df.loc[df['director'].isnull(), 'director'] = 'Unknown'
df.loc[df['cast'].isnull(), 'cast'] = 'Unknown'
df.loc[df['country'].isnull(), 'country'] = 'Unknown'
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
from sklearn.feature_extraction.text import TfidfVectorizer
df['soup'] = df['listed_in'] + ' ' + df['description'] + ' ' + df['title']
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df['soup'])
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
inertia_scores = []
k_range = range(2, 16) 

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia_scores.append(kmeans.inertia_)
    print(f"Completed for k={k}")
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_scores, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()

```

```python


optimal_k = 8
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(tfidf_matrix)
df['cluster'] = kmeans.labels_
print("DataFrame with cluster labels:")
print(df.head())
print("\nCluster sizes:")
print(df['cluster'].value_counts())
```

```python

cluster_number = 0

cluster_df = df[df['cluster'] == cluster_number]

print(f"--- Top 10 Genres in Cluster {cluster_number} ---")
print(cluster_df['listed_in'].value_counts().head(10))
print("\n")

print(f"--- Top 5 Countries in Cluster {cluster_number} ---")
print(cluster_df['country'].value_counts().head(5))
print("\n")

print(f"--- Sample of Titles in Cluster {cluster_number} ---")

print(cluster_df[['title', 'description']].sample(5))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


cluster_counts = df['cluster'].value_counts().sort_index()


plt.figure(figsize=(12, 7))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette='viridis')

plt.title('Number of Shows in Each Cluster', fontsize=16)
plt.xlabel('Cluster ID', fontsize=12)
plt.ylabel('Number of Shows', fontsize=12)
plt.show()
```

```python

from wordcloud import WordCloud
import matplotlib.pyplot as plt


for cluster_number in range(8):
    
    text = " ".join(show for show in df[df['cluster'] == cluster_number]['soup'])
    
    
    if text:
        
        wordcloud = WordCloud(
            background_color="white",
            colormap='RdPu', 
            collocations=False
        ).generate(text)

     
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.title(f'Most Common Words in Cluster {cluster_number}', fontsize=16)
        plt.show()
    else:
        print(f"Cluster {cluster_number} has no text to display.")
```
